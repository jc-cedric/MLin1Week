# Polynomial regression

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
```

```python
np.random.seed(0)

#dataset creation
x, y = make_regression(n_samples=100, n_features=1, noise=10)
y = y**2 # y doesn't no more variate linearly

#adding polynomial variables to our dataset
poly_features = PolynomialFeatures(degree=2, include_bias=False)
x = poly_features.fit_transform(x)

plt.scatter(x[:,0], y)
x.shape # x shape: 100 rows and 2 columns
```

```python
# model training
model = SGDRegressor(max_iter=1000, eta0=0.001)
model.fit(x,y)
print('Coeff R2 =', model.score(x, y))

plt.scatter(x[:,0], y, marker='o')
plt.scatter(x[:,0], model.predict(x), c='red', marker='+')
```
